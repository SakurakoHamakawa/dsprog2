```python
# 必要なライブラリをインポート
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup

# オプション設定（ヘッドレスにしたい場合など）
options = webdriver.ChromeOptions()
# options.add_argument("--headless")  # 画面を表示せずに実行したいとき

# ドライバを起動
driver = webdriver.Chrome(options=options)

# -------------------------
# STEP 1. 対象URLにアクセス
# -------------------------
url = "https://example.com"
driver.get(url)
time.sleep(2)  # ページ読み込み待ち（必要に応じて）

# -------------------------
# STEP 2. 必要なボタンや入力欄を操作
# -------------------------
# 例）クリックボタン
try:
    driver.find_element(By.CLASS_NAME, "some-class-name").click()
except Exception as e:
    print("クリックに失敗:", e)

# 例）チェックボックス
try:
    checkbox = driver.find_element(By.ID, "checkbox-id")
    if not checkbox.is_selected():
        checkbox.click()
except Exception as e:
    print("チェックボックス操作失敗:", e)

# 例）検索ボタン
try:
    driver.find_element(By.ID, "search-button-id").click()
except Exception as e:
    print("検索失敗:", e)

time.sleep(2)

# -------------------------
# STEP 3. ページから情報を取得（BeautifulSoupで解析）
# -------------------------
message = ""
click_count = 0
max_pages = 10  # ページ数上限（適宜変更）

while click_count < max_pages:
    soup = BeautifulSoup(driver.page_source, "html.parser")

    # 情報の抽出（例：商品一覧・求人・テーブルデータなど）
    items = soup.find_all("div", class_="item-class")  # 適切なタグとクラスを指定
    for item in items:
        try:
            name = item.find("span", class_="name").text.strip()
            price = item.find("span", class_="price").text.strip()
            message += f"{name} ／ {price}\n"
        except Exception as e:
            print("データ取得エラー:", e)
            continue

    # -------------------------
    # STEP 4. 「次へ」ボタンでページ移動
    # -------------------------
    try:
        next_button = driver.find_element(By.NAME, "next-button-name")  # 適切な要素を指定
        next_button.click()
        click_count += 1
        time.sleep(1)
    except Exception as e:
        print("次ページに進めませんでした:", e)
        break

# -------------------------
# STEP 5. 結果の出力と終了処理
# -------------------------
print(message)
driver.quit()
```
